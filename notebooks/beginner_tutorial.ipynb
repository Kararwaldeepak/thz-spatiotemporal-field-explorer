# Beginner Tutorial: Spatiotemporal THz Electric Field
This notebook runs the complete educational example step by step.

```python
from thz_field_analysis import calculate_scan_metrics

metrics = calculate_scan_metrics(stage_step_um=10, n_time_samples=201, delay_multiplier=2)
metrics
```

A 10 µm mechanical step in a retroreflector gives a 20 µm optical-path change. Dividing by the speed of light gives the time step.

```python
from thz_field_analysis import make_synthetic_vector_pulse

t_s, Ex, Ey = make_synthetic_vector_pulse(
    n_time_samples=201,
    stage_step_um=10,
    delay_multiplier=2,
)
Ex.shape, Ey.shape
```

The shape is `(time, y, x)`. Every time index is one camera image.

```python
from thz_field_analysis import analyze_field

analysis, arrays = analyze_field(Ex, Ey, dt_s=metrics.time_step_s)
analysis
```

```python
from thz_field_analysis import create_all_visuals

create_all_visuals(
    Ex, Ey, metrics, analysis, arrays,
    output_dir='outputs/notebook',
    max_spectrum_thz=3.0,
)
```
